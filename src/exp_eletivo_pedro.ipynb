```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['mathtext.fontset'] = 'stix'
```

```python
def seno_ang(x,y1,y2):
    ym = 0.5*(y1+y2)
    return (ym/2)/(np.sqrt(x**2 + (ym/2)**2))

def er_ym(er_y1,er_y2):
    return 0.5 * np.sqrt(er_y1**2 + er_y2**2)

def er_sin(ym,x,er_x,er_ym):
    Y = ym/2

    A = (Y*er_x)**2
    B = (x*er_ym)**2
    C = (x**2 + Y**2)**(3/2)
    return (x*(np.sqrt(A + B)))/(C)

def espessura_e_erro(A,er_A):
    lamb = 670
    er_lamb = 10

    d = (lamb/1000)/A

    er_d = d * np.sqrt((er_lamb/lamb)**2 + (er_A/A)**2)

    return d, er_d
```

### Cabelo do Lucas

```python
mL = np.array([1,2,3,4,5])
xL = (267.75-30)*10
y1L = np.array([53.55,84.40,114.90,145.00,172.6])
y2L = np.array([34.45,64.60,95.10,127.20,155.3])

er_xL = 1
er_y1L = np.array([0.05,0.05,0.05,0.05,0.5])
er_y2L = np.array([0.05,0.05,0.05,0.05,0.5])
```

```python
0.5*(y1L + y2L)
```

```python
er_ymL = er_ym(er_y1L,er_y2L)
```

```python
er_ymL
```

```python
er_sinL = er_sin(0.5*(y1L+y2L),xL,er_xL,er_ymL)
```

```python
er_sinL
```

```python
sinL = seno_ang(xL,y1L,y2L)
```

```python
sinL
```

```python
er_AL = 7.51899809854008e-06
AL = 0.00644073563562274

espessura_e_erro(AL,er_AL)
```

```python
X = np.linspace(1,5,1000)
f = 0.00644073563562274*X + 0.00279673492549869
```

```python
fig, ax = plt.subplots(1,1)
                                                             
ax.errorbar(
    mL,sinL,yerr=er_sinL, fmt="o", capsize=10, color = "orange",
    label = "Pontos experimentais com barras de incerteza"
    )
ax.plot(X,f,label = "Reta linearizada | $R^{2} = 0,9999$")

ax.set_ylabel(r"$\sin(\theta)$", fontsize = 12)
ax.set_xlabel(r"ordem dos máximos de difração", fontsize = 12)

ax.xaxis.set_major_formatter(FormatStrFormatter('%.1f'))
ax.yaxis.set_major_formatter(FormatStrFormatter('%.3f'))

ax.legend()
ax.grid()

plt.savefig('grafico_1_exp_eletivo.png', dpi=300, bbox_inches='tight')
```
